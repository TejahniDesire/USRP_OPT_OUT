```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%load_ext autoreload
%autoreload 2
```

```python
data = open("data.txt","r")
data.readline()
data = data.read()
# print(data.split("\n"))
data = data.split("\n")
data.remove(data[len(data) - 1])
x = []
y = []
stds = []
for i in range(len(data)):
    line = data[i].split('\t')
    x += [float(line[0])]
    y += [float(line[1])]
    stds += [float(line[2])]
    print(x[i],y[i])
    
x = np.array(x)
y = np.array(y)
stds = np.array(stds)

"Sorting____________________________________"

ysorted = np.array([xs for _,xs in sorted(zip(x,y))])
stdsorted = np.array([xs for _,xs in sorted(zip(x,stds))])
print(stdsorted)
x.sort()
    
```

```python
plt.errorbar(x,ysorted,stdsorted)
plt.show
```

```python
a,b,c = np.polynomial.polynomial.polyfit(x,ysorted,2,w=1/stdsorted)

def func(xs):
    return c * (xs ** 2) + b * xs + a

plt.errorbar(x,ysorted,stdsorted,label="Data")
plt.plot(x,func(x),label="Best Fit")
plt.legend()
```
